```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
file1 = "./loss1.txt"; file2 ="./loss2.txt"
ep = []
for i in range(1, 20000+1):
    if(i<=3 or i%500==0):
        ep.append(i)

ep = np.array(ep)
data1 = np.loadtxt(file1).T
data2 = np.loadtxt(file2).T

plt.clf()
fig = plt.figure(dpi=100)
plt.yscale('log')
plt.plot(ep,data1[0,:], label="1dim train")
plt.plot(ep,data1[1,:], label="1dim val")

plt.plot(ep,data2[0,:], label="2dim train")
plt.plot(ep,data2[1,:], label="2dim val")
plt.legend()
plt.xlabel("epoch")
plt.ylabel("error")
plt.show()
```
