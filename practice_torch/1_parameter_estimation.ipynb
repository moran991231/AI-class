```python
%matplotlib inline
import numpy as np
import torch
import matplotlib.pyplot as plt
torch.set_printoptions(edgeitems =2, linewidth =75)
```

```python
t_c =[0.5, 14.0, 15.0, 28.0, 11.0, 8.0, 3.0, -4.0, 6.0, 13.0, 21.0]
t_u =[35.7, 55.9, 58.2, 81.9, 56.3, 48.9, 33.9, 21.8, 48.4, 60.4, 68.4]
t_c = torch.tensor(t_c)
t_u = torch.tensor(t_u)
```

```python
plt.plot(t_c, t_u,'x')
```

```python
def model(t_u, w, b):
    return w*t_u+b 

def loss_fn(t_p, t_c):
    square_diffs = (t_p -t_c)**2
    return square_diffs.mean()
```

```python
w = torch.ones(())
b = torch.zeros(())

t_p = model(t_u, w, b)
t_p
```

```python
loss = loss_fn(t_p, t_c)
loss
```

```python
delta =0.1

loss_rate_of_change_w = \
    (loss_fn(model(t_u, w+delta,b), t_c) - loss_fn(model(t_u,w-delta, b), t_c)) / (2.0*delta)
```

```python
learning_rate =1e-2
w -= learning_rate * loss_rate_of_change_w
```

```python
loss_rate_of_change_b = \
    (loss_fn(model(t_u, w,b+delta),t_c) - loss_fn(model(t_u, w, b-delta), t_c)) / (2.0*delta)
b -= learning_rate * loss_rate_of_change_b
```

```python
def dloss_fn(t_p, t_c):
    dsq_diffs = 2*(t_p - t_c) / t_p.size(0)
    return dsq_diffs

def dmodel_dw(t_u, w, b):
    return t_u

def dmodel_db(t_u, w, b):
    return 1.0
```

```python
def grad_fn(t_u, t_c, t_p,w, b):
    dloss_dtp = dloss_fn(t_p, t_c)
    dloss_dw = dloss_dtp * dmodel_dw(t_u, w, b)
    dloss_db = dloss_dtp * dmodel_db(t_u, w, b)
    return torch.stack([dloss_dw.sum(), dloss_db.sum()])
```

```python
def training_loop(n_epoches, learning_rate, params, t_u, t_c, print_params = True):
    for epoch in range(1, n_epoches+1):
        w, b = params
        t_p = model(t_u, w, b) #1
        loss= loss_fn(t_p, t_c)#2
        grad = grad_fn(t_u, t_c, t_p, w,b)

        params -= learning_rate * grad

        if epoch in{1,2,3,10,11, 99,100,4000,5000}:
            print('Epoch %d, Loss %f'%(epoch, float(loss)))
            if print_params:
                print('    Params: ', params)
                print('    Grad  : ', grad)
        if(epoch in {4, 12,101}):    
             print('...')   
        
        if not torch.isfinite(loss).all():
            break 
    return params
```

```python
training_loop(n_epoches=100, learning_rate=1e-2, params=torch.tensor([1.0, 0.0]), t_u=t_u, t_c=t_c, print_params = True)
```

```python
training_loop(n_epoches=100, learning_rate=1e-4, params=torch.tensor([1.0, 0.0]), t_u=t_u, t_c=t_c, print_params = True)
```

```python
t_un = 0.1*t_u
training_loop(n_epoches=100, learning_rate=1e-2, params=torch.tensor([1.0, 0.0]), t_u=t_un, t_c=t_c, print_params = True)
```

```python
params = training_loop(5000, 1e-2, torch.tensor([1.0, 0.0]), t_un, t_c, False)
params
```

```python
%matplotlib inline
from matplotlib import pyplot as pyplot
t_p = model(t_un, *params) # model(t_un, params[0],params[1])

fig = plt.figure(dpi=600)
plt.xlabel("Temperature (Fahrenheit)")
plt.ylabel("Temperature (Celsius)")
plt.plot(t_u.numpy(), t_p.detach().numpy())
plt.plot(t_u.numpy(), t_c.numpy(), 'o')
# plt.savefig("temp_unknown_plot.png", format="png")
```
